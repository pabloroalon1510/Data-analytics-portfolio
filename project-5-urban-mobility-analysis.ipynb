### 2.2 Renombrar columnas

**üéØObjetivo:**
Estandarizar los nombres de columnas para evitar errores y facilitar la uni√≥n de los datasets.

**Instrucciones:**

- Cambia los nombres de las columnas para que tengan el formato `snake_case`.
    - `Country` ‚Üí `country`
    - `UpdateTimeUTC` ‚Üí `update_time_utc`
- Verifica que los cambios se hayan aplicado correctamente usando `.columns`.


````python
# 1. Funci√≥n unificada para evitar repetitividad
def estandarizar_nombres(df):
    """Limpia nombres: min√∫sculas, quita espacios, par√©ntesis y s√≠mbolos."""
    df.columns = [
        col.lower().strip()
        .replace(' ', '_')
        .replace('/', '_')
        .replace('(', '')
        .replace(')', '')
        .replace('%', 'pct')
        for col in df.columns
    ]
    return df

# 2. Aplicar limpieza base a ambos
traffic = estandarizar_nombres(traffic)
eco = estandarizar_nombres(eco)

# 3. Mapeo espec√≠fico para Traffic (incluyendo 'country' y nombres compuestos)
# Aqu√≠ corregimos expl√≠citamente las columnas que no se separan solas
rename_dict = {
    'country': 'country', # Se asegura de estar en min√∫sculas
    'city': 'city',
    'updatetimeutc': 'update_time_utc',
    'jamsdelay': 'jams_delay',
    'trafficindexlive': 'traffic_index_live',
    'jamslengthinkms': 'jams_length_kms',
    'jamscount': 'jams_count',
    'trafficindexweekago': 'traffic_index_week_ago',
    'updatetimeutcweekago': 'update_time_utc_week_ago',
    'traveltimeliveper10kmsmins': 'travel_time_live_per_10kms_mins',
    'traveltimehistoricper10kmsmins': 'travel_time_hist_per_10kms_mins',
    'minsdelay': 'mins_delay'
}

traffic = traffic.rename(columns=rename_dict)

# 4. Verificaci√≥n profesional (limpia y concisa)
print("Estandarizaci√≥n de columnas completada.")
print(f"Columnas Traffic: {traffic.columns.tolist()[:5]}...") # Mostramos inicio para verificar 'country'
print(f"Columnas Economy: {eco.columns.tolist()[:5]}...")
````

### 2.3 Corregir formatos num√©ricos y de fecha

**üéØObjetivo:**
Asegurar que las columnas de fechas y valores num√©ricos est√©n en formatos correctos para permitir an√°lisis, c√°lculos y comparaciones precisas.

**Instrucciones:**

- Convierte las columnas de fecha de `traffic` a formato `datetime`. Haz el cambio a prueba de errores.
- En el dataset `eco`, limpia los valores num√©ricos:
    - En `city_gdp_capita`: elimina separadores de miles (`.`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `unemployment_pct`: elimina el s√≠mbolo de porcentaje (`%`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `population_m`: reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
- Finalmente, crea una nueva columna llamada `population` multiplicando `population_m` por 1,000,000 para obtener la poblaci√≥n total.


<details>
<summary>Haz clic para ver la pista</summary>

````python
# Convertir fechas en traffic (Ahora s√≠ encontrar√° 'update_time_utc')
traffic['update_time_utc'] = pd.to_datetime(traffic['update_time_utc'], errors='coerce')
traffic['update_time_utc_week_ago'] = pd.to_datetime(traffic['update_time_utc_week_ago'], errors='coerce')

# Limpiar y convertir columnas num√©ricas en eco
eco['city_gdp_capita'] = eco['city_gdp_capita'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
eco['unemployment_pct'] = eco['unemployment_pct'].astype(str).str.replace('%', '', regex=False).str.replace(',', '.', regex=False).astype(float)
eco['population_m'] = eco['population_m'].astype(str).str.replace(',', '.', regex=False).astype(float)

# Calcular poblaci√≥n absoluta
eco['population'] = eco['population_m'] * 1000000

# Verificar
eco.info()
````

### 3.1 Extraer columna a√±o y filtrar 2024

**üéØObjetivo**
Identificar el a√±o de cada registro y mantener solo los registros del 2024.

**Intrucciones**

## üß©Paso 4: Analizar y resumir datos de movilidad

Como el dataset de tr√°fico contiene **m√∫ltiples registros por ciudad**. En esta parte, calcular√°s los promedios anuales por ciudad para simplificar el an√°lisis y obtener una visi√≥n m√°s clara de las tendencias generales.

### 4.1 Calcular promedios de tr√°fico por ciudad

**üéØObjetivo:**
Obtener una vista consolidada del tr√°fico promedio por ciudad y a√±o, para analizar patrones generales sin depender de datos diarios.

**Instrucciones**

- Agrupa los datos por `city`, `country` y `year`.
- Calcula el promedio **solo de las m√©tricas de tr√°fico m√°s relevantes**: como `jams_delay`, `traffic_index_live`, `jams_length_kms`, `jams_count`, `mins_delay`, y tiempos de viaje (`travel_time_live_per_10kms_mins` y `travel_time_hist_per_10kms_mins`).
- Guarda el resultado como `traffic_city_year_2024`, mant√©n las columnas como variables (no √≠ndices).


<details>
<summary>Haz clic para ver la pista</summary>
Usa ".agg()" para aplicar funciones de promedio. Al final, reinicia el √≠ndice para mantener las columnas de la agrupaci√≥n como variables (no √≠ndices).

````python
# Calcular los promedios de trafico por ciudad, pa√≠s y a√±o
traffic_city_year_2024 = traffic_2024.groupby(['city', 'country', 'year']).agg({
    'jams_delay': 'mean',
    'traffic_index_live': 'mean',
    'jams_length_kms': 'mean',
    'jams_count': 'mean',
    'mins_delay': 'mean',
    'travel_time_live_per_10kms_mins': 'mean',
    'travel_time_hist_per_10kms_mins': 'mean'
}).reset_index()

# Mostrar resultado
display(traffic_city_year_2024.head())
````

````python
# Ordenar por el retraso promedio (jams_delay) de mayor a menor
ranking_trafico = traffic_city_year_2024.sort_values(by='jams_delay', ascending=False)

# Mostrar las 10 ciudades con peor tr√°fico
display(ranking_trafico.head(10))
````

## üß©Paso 5: Unir movilidad y econom√≠a

Combinar datasets te permite analizar c√≥mo se relacionan los indicadores econ√≥micos con los de movilidad.

### 5.1 Unir tr√°fico (tabla principal) con indicadores econ√≥micos

**üéØObjetivo:**
Combinar la informaci√≥n de tr√°fico y econom√≠a en un solo DataFrame para analizar c√≥mo las condiciones econ√≥micas se relacionan con la movilidad urbana.

**Instrucciones**
- Selecciona solo las **columnas relevantes** de cada dataset (por ejemplo, variables clave de tr√°fico y de econom√≠a).
- Usa `.copy()` al crear subconjuntos para evitar modificar el dataset original.
- Une ambos DataFrames y define como **claves de uni√≥n** a `city` y `year`.
- Mant√©n solo las ciudades y a√±os presentes en ambos datasets.
- Guarda el resultado en una nueva variable llamada `merged` y muestra las primeras 5 filas.


<details>
<summary>Haz clic para ver la pista</summary>
Aplica una uni√≥n de tipo "inner" para mantener las ciudades y a√±os presentes en ambos datasets.

````python
# 5.1 Seleccionar solo las columnas necesarias para el an√°lisis final
left_cols = ['city', 'year', 'jams_delay', 'traffic_index_live', 'mins_delay']
right_cols = ['city', 'year', 'city_gdp_capita', 'unemployment_pct', 'population']

# 5.2 Aplicar la uni√≥n de tipo "inner"
# Esto descarta ciudades que no tengan datos en ambos archivos
merged = pd.merge(
    traffic_city_year_2024[left_cols], 
    eco_2024[right_cols], 
    on=['city', 'year'], 
    how='inner'
)

# 5.3 Mostrar el resultado final ordenado por congesti√≥n
print("Resultado de la uni√≥n (Inner Join):")
display(merged.sort_values(by='jams_delay', ascending=False).head(10))
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Unir y Limpiar (Agrupamos por ciudad para tener valores √∫nicos de PIB y Tr√°fico)
merged = pd.merge(traffic, eco, on='city', how='inner')
merged_grouped = merged.groupby('city').agg({
    'jams_delay': 'mean',
    'city_gdp_capita': 'mean'
}).reset_index()

# 2. Preparar el Top 20 para el ranking de congesti√≥n
top_cities = merged_grouped.sort_values('jams_delay', ascending=False).head(20)

# 3. Configurar el lienzo con 2 gr√°ficos (Subplots)
fig, (ax_hist, ax_bar) = plt.subplots(1, 2, figsize=(22, 9))

# --- GR√ÅFICO 1: Histograma de la Econom√≠a (City GDP Capita) ---
sns.histplot(data=merged_grouped, x='city_gdp_capita', kde=True, ax=ax_hist, color='#2a9d8f')
ax_hist.set_title('1. Distribuci√≥n de la Econom√≠a (PIB per c√°pita)', fontsize=14, fontweight='bold')
ax_hist.set_xlabel('PIB per c√°pita (USD)')
ax_hist.set_ylabel('Frecuencia (N¬∞ de Ciudades)')

# --- GR√ÅFICO 2: Ranking con Doble Eje (Tr√°fico vs PIB) ---
# Barras para el Tr√°fico
sns.barplot(data=top_cities, x='city', y='jams_delay', ax=ax_bar, color='salmon', alpha=0.7)
ax_bar.set_ylabel('Jams Delay (Minutos)', fontsize=12, fontweight='bold', color='darkred')

# Eje secundario para el PIB (L√≠nea)
ax_pib = ax_bar.twinx()
sns.lineplot(data=top_cities, x='city', y='city_gdp_capita', ax=ax_pib, color='royalblue', marker='o', linewidth=2.5)
ax_pib.set_ylabel('PIB per c√°pita (USD)', fontsize=12, fontweight='bold', color='darkblue')

# --- TU TIP: Rotaci√≥n de 90 grados ---
ax_bar.set_xticks(range(len(top_cities['city'])))
ax_bar.set_xticklabels(top_cities['city'], rotation=90)

ax_bar.set_title('2. Top 20: Tr√°fico vs Riqueza', fontsize=14, fontweight='bold')

# 4. Ajuste final de dise√±o
plt.tight_layout()
plt.show()
````
